Visitor mobility rates via different locations per country during COVID-19 outbreak.

###Initial JSON viewing + cleanup

```python
import json
import pandas as pd
import requests


covid_mobility =  'https://tnguyen.greenriverdev.com/SDEV121/covid_mobility.json'

response = requests.get(covid_mobility)
json_data = response.json()
json_data


df = pd.json_normalize(json_data, sep="_")
df
```

Renaming columns to read easier

```python
import pandas as pd

df = df.rename(columns={'Location_Retail':'Retail',
                        'Location_Grocery':'Grocery',
                        'Location_Parks':'Parks',
                        'Location_Public transport':'Public Transport',
                        'Location_Workplace':'Workplace',
                        'Location_Residential':'Residential'})
```

The initial DataFrame includes over 130 countries with individual columns based on how many days are counted from initial outbreak.

This makes the DataFrame super redundant, as its printing rows per individual day.

To solve this issue, we noticed a trend. Each country stops at 226 days, meaning that the DataFrame covers countries towards 226 days since outbreak.
(Confirmed with code:
df.tail() and converting table to interactive table.
)
We split the initial DataFrame into 3 separate DataFrames:



1. df_max contains the mobility rates at the peak day since outbreak.
2. df_med contains the mobility rates at the median day since outbreak.
3. df_min contains the mobility rates at the minimum day since outbreak.

Using these 3 DataFrames, we can compare mobility rates more directly using various calculations.
*All DataFrame values are rounded.*

*Note: Because some countries have days that aren't noted (ex: Afghanistan only has 132 rows of data, whereas Angola has 180 rows implying that some data within Afghanistan were missed.), some countries are under a different median value (calculated separately).*

###Sorted DataFrames

```python
#finds sum for all countrys' days since outbreak
df.groupby('Country')['Days since outbreak'].sum() # 24570 sum for countries with 180 days counted

#finding what countries aren't up to the 180 day sum
df.groupby('Country')['Days since outbreak'].sum() == 24570

#Afghanistan, Georgia, Serbia median needs to be calculated separately
```

Querying the alike median for the three countries

```python
afg = df[(df['Country'] == 'Afghanistan') & (df['Days since outbreak'] == 114)]
afg

geo = df[(df['Country'] == 'Georgia') & (df['Days since outbreak'] == 114)]

ser = df[(df['Country'] == 'Serbia') & (df['Days since outbreak'] == 114)]
```

DataFrame querying the max days.

```python
max = (df['Days since outbreak'].max()) #find peak days
df_max = df[df['Days since outbreak']==max] #create df_peak where days query to "max" variable
df_max.set_index('Country',inplace=True) #set index
df_max
```

DataFrame querying the median days.

```python
median = (df['Days since outbreak'].median()) #find median days
df_med = df[df['Days since outbreak']==median] #create df_peak where days query to "median" variable
df_med.set_index('Country',inplace=True) #set index

df_med.drop(index=['Afghanistan','Georgia','Serbia'])

#adding back rows with proper median query

#Afghanistan
df_med.loc['Afghanistan'] = afg.iloc[0]

#Georgia
df_med.loc['Georgia'] = geo.iloc[0]


#Serbia
df_med.loc['Serbia'] = ser.iloc[0]


df_med
```

DataFrame querying minimum days.

```python
minimum = (df['Days since outbreak'].min()) #find overall minimum days
df_min = df[df['Days since outbreak']==minimum] #create df_minimum where days query to "minimum" variable
df_min.set_index('Country',inplace=True) #set index
df_min
```

```python
#to find the highest and lowest
ny_max_cases = ny_max.sort_values('cases', ascending = False)
ny_max_cases
```
